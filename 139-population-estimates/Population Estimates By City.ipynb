Similar to the population estimates from 2010 tract, but this uses data that is already at the city granularity. 

```python
import pandas as pd
import sqlite3
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

# 2013 series estimates to 2050, by jurisdiction (cities) extracted and downloaded through the
# web interface. 
sandag2050_h = pd.read_excel('http://ds.civicknowledge.org/sandag.org/pop-projections/Housing_Forecast_2050_Jurisdictions.xls')
sandag2050_h.set_index(['Year','Area', 'ID'], inplace = True)

sandag2050_ag = pd.read_excel('http://ds.civicknowledge.org/sandag.org/pop-projections/AgeGender_Forecast_2050_Jurisdictions.xls')
sandag2050_ag.set_index(['Year','Area', 'ID'], inplace = True)

sandag2050_r = pd.read_excel('http://ds.civicknowledge.org/sandag.org/pop-projections/Race_Forecast_2050_Jurisdictions.xls')
sandag2050_r.set_index(['Year','Area', 'ID'], inplace = True)
```

```python
sandag2050 = sandag2050_h.join(sandag2050_ag).join(sandag2050_r).reset_index()
sandag2050.head()
```

```python
def calc_rates(s):
    """Return a series of extrapolations, from the three years 2012, 2020 and 2035"""
    x = np.append(np.linspace(s[2008], s[2015],8),np.linspace(s[2015], s[2020],6)[1:])
    x = np.append(x, np.linspace(s[2020], s[2035],16)[1:])
    return pd.Series(x, index=range(2008,2036)).astype(int)

def extrapolate_col(df, col_name):
    """Create a new DataFrame with the extrapolations of a named column. The values are stacked, so
    multiple extrapolations can be joined. """
    dfp = df.pivot(index='Area', columns='Year', values=col_name)
    
    r =  dfp.apply( calc_rates, axis=1)
   
    r = r.stack()
    
    x =  pd.DataFrame(r)
    x.index.names = ['city','year' ]
    x.columns =  [col_name]

    return x
 
```

```python
r = extrapolate_col(sandag2050, 'Population' )

for col_name in sandag2050.columns[4:]:
    r = r.join(extrapolate_col(sandag2050, col_name ))

r = r.reset_index()
r = r[r.year >= 2015]
r.to_csv('2008_series.csv')
r
```

```python

r2 = r.reset_index().sort(['city', 'year'])

diffs = r2[['city','0-4 M','0-4 F', 'Population']].groupby(['city']).transform(lambda x: (x-x.shift(1))/x.shift(1) )
diffs.columns = [ x+' diff' for x in diffs.columns]
kids = r2.join(diffs)
```

```python
all_kids = r.reset_index()[['city','year','0-4 M','0-4 F','5-9 M', '5-9 F']]
all_kids['0-9'] = all_kids['0-4 M'] + all_kids['0-4 F'] + all_kids['5-9 M'] + all_kids['5-9 F']
kids = all_kids[['city','year','0-9']].sort(['city', 'year'])
diffs = kids[['city','0-9']].groupby(['city']).transform(lambda x: (x-x.shift(1))/x.shift(1) )
diffs.columns = ['0-9 change']
diffs['0-9 change'] = pd.Series.round(diffs['0-9 change'] *100 , 2)
joined = kids.join(diffs)
```

```python
jc = joined.pivot(index='year', columns='city', values='0-9 change')
jc.columns = [ c+" %" for c in jc.columns]

jv = joined.pivot(index='year', columns='city', values='0-9')

jvc = jv.join(jc)
jvc = jvc.sort_index(axis=1)
jvcs  = jvc.loc[2015:2025]
jvcs
```
