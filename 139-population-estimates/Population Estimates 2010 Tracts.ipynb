# San Diego Population Estimates, 2012 to 2035

Extrapolations of the 2012 SANDAG population estimates to consecutive years. This uses a special extract of the SANDAG data warehouse to 2010 tracts. 

[Issue Ticket 139](http://redmine.civicknowledge.com/issues/139)


```python
import pandas as pd
import sqlite3
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

host  = "http://localhost:8081"

url = host+'/warehouses/skwYid7pZN/extracts/forecast13.csv'

df = pd.read_csv(url)
df.head()
```

```python

def calc_rates(s):
    """Return a series of extrapolations, from the three years 2012, 2020 and 2035"""
    return pd.Series(np.append(np.linspace(s[2012], s[2020],9), 
                               np.linspace(s[2020], s[2035],16)[1:]), index=range(2012,2036)).astype(int)

def extrapolate_col(df, col_name):
    """Create a new DataFrame with the extrapolations of a named column. The values are stacked, so
    multiple extrapolations can be joined. """
    dfp = df.pivot(index='c03d01002_censustract2010', columns='c03d01003_year', values=col_name)
    
    r =  dfp.apply( calc_rates, axis=1)
   
    r = r.stack()
    
    x =  pd.DataFrame(r)
    x.index.names = ['tract','year' ]
    x.columns =  [col_name]

    return x
 
r = extrapolate_col(df, 'c03d01004_pop' )

for col_name in df.columns[4:]:
    r = r.join(extrapolate_col(df, col_name ))

r.to_csv('popest_single_years.csv')
r.head()
```

Link in the ACS geofile though the tract number

```python
act = pd.read_csv(host+"/warehouses/skwYid7pZN/extracts/cities_tracts.csv")
act.head()
```

```python
acts = act[['c02G0100f_tract','c03l01003_city']]
# since the r dataframe has a multiindex, this table must have an 
# index of the same name to be able to join w/o an 'on' argument
acts.columns = [ 'tract','city'] 
acts.set_index('tract',inplace = True)
acts.head()
```

```python
cities = r.join(acts).reset_index().groupby(['year','city']).sum()
SndSAN = cities.xs('SndSAN',level='city')
```
